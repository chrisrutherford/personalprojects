# Data visualation with Pokémon stats
##### source: https://www.kaggle.com/abcsds/pokemon/

### Importing the dataset and main libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
pokemon=pd.read_csv("Pokemon.csv")
```

```python
pokemon
```

```python
pokemon=pokemon.drop_duplicates("#",keep="first")
```

```python
np.sum(pokemon.isna())
```

```python
pokemon.describe()
```

```python
plt.figure(figsize=(12,6))
sns.distplot(pokemon['Total'],bins=15, kde=False,
            hist_kws={"alpha": 1, "color": "#aa0000"})
plt.suptitle("Pokémon Overall Stat Distribution",fontsize=20)
plt.title("Min: {} Max: {} Avg: {}".format(np.min(pokemon['Total']),
                                           np.max(pokemon['Total']),
                                           round(np.mean(pokemon['Total']),2)),
                                           fontsize=16)
plt.xlabel("Sum of Base Stats",fontsize=15)
plt.ylabel("Frequency",fontsize=15)
plt.show()
```

```python
pokemon_dragon=pokemon[(pokemon['Type 1']=="Dragon") | (pokemon['Type 2']=="Dragon")]
```

```python
pokemon_dragon.head()
```

```python
plt.figure(figsize=(12,6))
sns.distplot(pokemon_dragon['Total'],bins=15, kde=False,
            hist_kws={"alpha": 1, "color": "#4924A1"})
plt.suptitle("Dragon Type Pokémon Overall Stat Distribution",fontsize=20)
plt.title("Min: {} Max: {} Avg: {}".format(np.min(pokemon_dragon['Total']),
                                           np.max(pokemon_dragon['Total']),
                                           round(np.mean(pokemon_dragon['Total']),2)),
                                           fontsize=16)
plt.xlabel("Sum of Base Stats",fontsize=15)
plt.ylabel("Frequency",fontsize=15)
plt.show()
```
